```python
# When used as a python module (as we are doing) TWINT can be configured with various function calls
# (search term, limit results, etc). Below we will configure a basic search for a specific users last 10 tweets.
```

```python
# Get the last 10 tweets from a user and run
# Note the limit doesn't always work as expected (it may be counting tweets to limit but also displaying replies?)
c = twint.Config()

c.Username = 'jack'
c.Limit = 10

twint.run.Search(c)
```

```python
# TWINT has a large number of configuration options
# https://github.com/twintproject/twint/wiki/Configuration
# Some of the most useful are since, until and output
```

```python
# Get a specific users tweets between dates shown and output to a csv file
# Check Jupyter homepage tab for csv file. Can be read into other python programs, Notebooks and cells.
c = twint.Config()

c.Username = 'jack'
c.Since = '2020-05-20 00:00:00'
c.Until = '2020-05-28 00:00:00'
c.Output_csv = True
c.Output = 'jack-tweets-20-28-may-2020.csv'

twint.run.Search(c)
```

```python
# lets make the information easier to read and manipulate inside Jupyter by using Pandas dataframes (we imported pandas earlier)
# Display jacks tweets between 20th 28th May 2020

c = twint.Config()

c.Username = 'jack'
c.Since = '2020-05-20 00:00:00'
c.Until = '2020-05-28 00:00:00'
c.Pandas = True

twint.run.Search(c)
```
